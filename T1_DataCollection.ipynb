# Python Script to Download Posters

```python
import pandas as pd
import requests
import os
import re
import time
from tqdm import tqdm

# Load dataset
df = pd.read_csv('movies_train.csv')
os.makedirs('posters', exist_ok=True)

API_KEY = '5530ad51b9026b431326f51af1a07d3a'

def poster_exists(movie_id):
    return os.path.exists(f"posters/{movie_id}.jpg")

def get_movie_poster(title, movie_id):
    if poster_exists(movie_id):
        return True
    
    clean_title = re.sub(r'\s*\(\d{4}\)', '', title).strip()
    
    try:
        search_url = "https://api.themoviedb.org/3/search/movie"
        params = {'api_key': API_KEY, 'query': clean_title}
        response = requests.get(search_url, params=params, timeout=10)
        
        if response.status_code == 200:
            results = response.json()['results']
            if results:
                poster_path = results[0].get('poster_path')
                if poster_path:
                    poster_url = f"https://image.tmdb.org/t/p/w500{poster_path}"
                    poster_response = requests.get(poster_url, timeout=10)
                    
                    if poster_response.status_code == 200:
                        with open(f"posters/{movie_id}.jpg", 'wb') as f:
                            f.write(poster_response.content)
                        print(f"✓ Downloaded: {title}")
                        return True
                    else:
                        print(f"✗ Download failed: {title}")
                else:
                    print(f"✗ No poster: {title}")
            else:
                print(f"✗ Not found: {title}")
        else:
            print(f"✗ API error: {title}")
    except Exception as e:
        print(f"✗ Error: {title} - {str(e)}")
    
    return False

# Download only missing posters
print("Downloading missing posters...")
existing = sum(1 for mid in df['movieId'] if poster_exists(mid))
success = existing

for _, row in tqdm(df.iterrows(), total=len(df)):
    if not poster_exists(row['movieId']):
        if get_movie_poster(row['title'], row['movieId']):
            success += 1
        time.sleep(0.1)

print(f"\nFinal: {success}/{len(df)} posters")
```

# Enhanced Search for Missing Movies

```python
import pandas as pd
import requests
import os
import re
import time

# Load missing movies
missing_df = pd.read_csv('missing_posters.csv') if os.path.exists('missing_posters.csv') else None

if missing_df is None:
    # Create missing list from what wasn't downloaded
    downloaded = set(int(f.split('.')[0]) for f in os.listdir('posters') if f.endswith('.jpg'))
    all_movies = set(pd.read_csv('movies_train.csv')['movieId'])
    missing_ids = all_movies - downloaded
    
    missing_data = []
    df = pd.read_csv('movies_train.csv')
    for movie_id in missing_ids:
        movie_row = df[df['movieId'] == movie_id].iloc[0]
        missing_data.append({'movieId': movie_id, 'title': movie_row['title']})
    
    missing_df = pd.DataFrame(missing_data)
    missing_df.to_csv('missing_posters.csv', index=False)

print(f"Missing posters: {len(missing_df)}")

API_KEY = '5530ad51b9026b431326f51af1a07d3a'

def enhanced_search(title, movie_id):
    """Enhanced search with multiple strategies"""
    original_title = title
    clean_title = re.sub(r'\s*\(\d{4}\)', '', title).strip()
    year = re.search(r'\((\d{4})\)', title)
    year = year.group(1) if year else None
    
    # Try different search strategies
    strategies = [
        {'query': clean_title, 'year': year},  # Original with year
        {'query': clean_title},  # Original without year
        {'query': clean_title.replace('The ', '')},  # Without "The"
        {'query': clean_title.replace('A ', '')},  # Without "A"
        {'query': clean_title.split(':')[0]},  # Without subtitle
        {'query': clean_title.split('-')[0]},  # Without hyphen part
    ]
    
    for strategy in strategies:
        try:
            search_url = "https://api.themoviedb.org/3/search/movie"
            params = {'api_key': API_KEY, 'query': strategy['query']}
            if strategy.get('year'):
                params['year'] = strategy['year']
            
            response = requests.get(search_url, params=params, timeout=10)
            
            if response.status_code == 200:
                results = response.json()['results']
                if results:
                    poster_path = results[0].get('poster_path')
                    if poster_path:
                        poster_url = f"https://image.tmdb.org/t/p/w500{poster_path}"
                        poster_response = requests.get(poster_url, timeout=10)
                        
                        if poster_response.status_code == 200:
                            with open(f"posters/{movie_id}.jpg", 'wb') as f:
                                f.write(poster_response.content)
                            print(f"✓ Found: {original_title}")
                            return True
        except:
            continue
    
    print(f"✗ Still missing: {original_title}")
    return False

# Try enhanced search on missing movies
print("Trying enhanced search on missing movies...")
new_success = 0

for _, row in missing_df.iterrows():
    if enhanced_search(row['title'], row['movieId']):
        new_success += 1
    time.sleep(0.15)  # Be gentle with API

print(f"Newly found: {new_success}/{len(missing_df)}")
```

# Downloading missed posters images with (high quality)

```python
import requests
import pandas as pd
import os
import re
import time
from io import BytesIO
from PIL import Image, ImageOps
from bs4 import BeautifulSoup

def find_posters_google_clean(csv_file, download_folder="missed_clean_highres"):
    """
    Improved version that filters for real movie posters and removes irrelevant images.
    """
    df = pd.read_csv(csv_file)
    os.makedirs(download_folder, exist_ok=True)

    headers = {
        "User-Agent": (
            "Mozilla/5.0 (Windows NT 10.0; Win64; x64) "
            "AppleWebKit/537.36 (KHTML, like Gecko) "
            "Chrome/118.0 Safari/537.36"
        )
    }

    poster_keywords = ["poster", "movie", "film", "imdb", "tmdb", "cinema"]
    exclude_keywords = ["logo", "banner", "fanart", "avatar", "profile", "background", "wallpaper"]

    success_count = 0

    for index, row in df.iterrows():
        movie_id, title = row["movieId"], row["title"]
        clean_title = title.split("(")[0].strip() + " movie poster"
        print(f"🎬 Searching: {clean_title}")

        try:
            query = clean_title.replace(" ", "+")
            search_url = f"https://www.google.com/search?tbm=isch&q={query}"

            response = requests.get(search_url, headers=headers, timeout=10)
            if response.status_code != 200:
                print(f"❌ Search failed ({response.status_code}) for {clean_title}")
                continue

            soup = BeautifulSoup(response.text, "html.parser")

            # Extract possible full-size URLs
            image_urls = set()
            for img in soup.find_all("img"):
                for attr in ["data-iurl", "data-src", "src"]:
                    url = img.get(attr)
                    if url and url.startswith("http") and "encrypted-tbn0" not in url and "gstatic" not in url:
                        image_urls.add(url)

            # Extract URLs from JSON patterns
            matches = re.findall(r'\"(https?://[^"]*\.(?:jpg|jpeg|png))\"', response.text)
            for m in matches:
                if "encrypted-tbn0" not in m and "gstatic" not in m:
                    image_urls.add(m)

            if not image_urls:
                print(f"❌ No image URLs found for {clean_title}")
                continue

            # Filter by relevance
            filtered = [
                url for url in image_urls
                if any(k in url.lower() for k in poster_keywords)
                and not any(k in url.lower() for k in exclude_keywords)
            ]

            if not filtered:
                filtered = list(image_urls)  # fallback

            # Try downloading filtered URLs
            for img_url in filtered[:10]:
                try:
                    img_response = requests.get(img_url, headers=headers, timeout=10)
                    if img_response.status_code != 200 or len(img_response.content) < 80_000:
                        continue

                    # Load image
                    image = Image.open(BytesIO(img_response.content)).convert("RGB")
                    w, h = image.size
                    aspect = h / w if w > 0 else 0

                    # Skip non-poster shapes (too wide)
                    if aspect < 1.2:
                        continue

                    # Auto-crop border (if white/black padding)
                    image = ImageOps.crop(image, border=5)
                    bg = Image.new("RGB", image.size, image.getpixel((0, 0)))
                    diff = ImageOps.invert(ImageOps.autocontrast(image))
                    image = ImageOps.crop(diff, border=10)

                    # Save final image
                    filename = f"{movie_id}.jpg"
                    filepath = os.path.join(download_folder, filename)
                    image.save(filepath, "JPEG", quality=90)

                    print(f"✅ Saved clean poster: {filename} ({w}x{h})")
                    success_count += 1
                    break
                except Exception:
                    continue
            else:
                print(f"❌ No valid poster found for {clean_title}")

        except Exception as e:
            print(f"⚠️ Error: {e}")

        time.sleep(1.5)

    print(f"\n📊 Summary: {success_count}/{len(df)} clean posters downloaded")
    return success_count


# Example usage
find_posters_google_clean("missing_posters_2.csv")
```
